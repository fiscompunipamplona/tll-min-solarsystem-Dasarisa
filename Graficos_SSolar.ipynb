```python
import numpy as np #Imports Python mathematical functions library
import matplotlib.pyplot as plt #Imports plot library
```

```python
cos = np.cos
pi = np.pi

q_Mercu = 0.313  #Ua
e_Mercu = 0.2056
theta = np.linspace(0,2*pi, 360)
r_Mercu = (q_Mercu*(1-e_Mercu**2))/(1+e_Mercu*cos(theta))
plt.polar(theta, r_Mercu)

#print(np.c_[r,theta])

plt.show()
```
